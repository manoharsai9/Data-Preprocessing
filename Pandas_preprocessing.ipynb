# Treating Missing Values

```python
#where we are having null values and how they sum up to?
dataFrame.isnull().sum()
```

```python
#drop the entire two rows missing Embarked entry because they are very very limited compared to cabina and age missing values

dataFrame = dataFrame.dropna(subset=['Embarked'])
```

```python
dataFrame.shape
```

```python
#cabin missing values are very large, so drop the entire column

dataFrame = dataFrame.drop('Cabin', axis=1)
```

```python
#age missing values are filled with mean of ages
mean_of_ages = np.mean(dataFrame['Age'])
dataFrame = dataFrame.fillna(value = {'Age' : mean_of_ages})
```
